# Pandas란?

### 데이터를 분석 및 조작하기 위한 라이브러리
##### 서로 다른 여러 유형의 데이터를 공통의 포맷으로 정리
##### 행과 열로 이루어진 2차원 구조의 데이터프레임에서 자주 사용됨
### 데이터프레임(DataFrame)과 시리즈(Series)라는 데이터 타입을 제공
##### 데이터프레임: 엑셀 시트와 같은, 테이블 형태의 구조 (2차원 배열)
##### 시리즈: 데이터프레임에서 한 개의 열에 해당 (column)

```python
import pandas as pd
```

```python
# csv 파일 읽기
CCTV_Seoul = pd.read_csv('01. CCTV_in_Seoul.csv', encoding = 'utf-8')
CCTV_Seoul.head() # head()명령은 pandas 데이터의 첫 5행만 보여달라는 뜻
```

```python
CCTV_Seoul.columns
```

```python
CCTV_Seoul.columns[0]
```

```python
# 제목 변경 
CCTV_Seoul.rename(columns = {CCTV_Seoul.columns[0] : '구별'}, inplace = True)
CCTV_Seoul.head()
```

```python
# 소계를 기준으로 데이터 정렬
CCTV_Seoul.sort_values(by='소계', ascending=True).head(5)
```

```python
# 증가율 계산: 2014~2016년 CCTV 수 더하고, 2013년 이전 CCTV 수로 나눔
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년']) + (CCTV_Seoul['2015년']) + \
                           (CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head(5)
```

```python
# 엑셀 파일 읽기
pop_Seoul = pd.read_excel('01. population_in_Seoul.xls')
pop_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('01. population_in_Seoul.xls',
                         header = 2, # 엑셀의 세번째 줄부터 읽어옴
                         usecols = 'B, D, G, J, N') # B,D,G,J,N열을 읽어옴
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별',
                          pop_Seoul.columns[1] : '인구수',
                          pop_Seoul.columns[2] : '한국인',
                          pop_Seoul.columns[3] : '외국인',
                          pop_Seoul.columns[4] : '고령자'}, inplace = True)
pop_Seoul.head()
```

```python
# 비율 계산
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
pop_Seoul.head()
```

```python
# 데이터 합치기
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')
data_result.head()
```

```python
# 데이터 삭제
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result
```

```python
# Index 기준을 변경
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
# 그래프로 분석 - matplotlib는 한글 지원이 안되기 때문에 폰트 변경 필요
import platform
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False
# Apple
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
# Windows
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else: 
    print('Unknown system... Sorry.')
```

```python
data_result['소계'].sort_values().plot(kind='barh',
                                    grid=True, figsize=(10,10))
plt.show()
```

```python
# 인구 대비 CCTV 비율 계산, 정렬
data_result['CCTV비율'] = data_result['소계'] / data_result['인구수'] * 100

data_result['CCTV비율'].sort_values().plot(kind='barh',
                                        grid=True, figsize=(10,10))
plt.show()
```

```python
# CCTV와 인구수 상관관계 그래프
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
import numpy as np
# polyfit 함수를 이용해 예측모델(!차월 함수) fp1의 계수 만들기
fp1 =np.polyfit(data_result['인구수'], data_result['소계'], )

# 만들어낸 예측모델을 이용한 그래프 그리기
f1 = np.poly1d(fp1)
# 10만~70만까지 총 100개의 구간으로 나눠서 표현
fx = np.linspace(100000, 700000, 100)
# 오차 값 계산 및 오차 컬럼 생성(추가)
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))
# 오차 값을 기준으로 내림차순으로 정렬
df_sort = data_result.sort_values(by='오차', ascending=False)
# 0번째~4번째 행까지의 값을 출력
df_sort.head()
```

```python
# 설득력 있는 자료 만들기 - 멀리 있는 구 이름 표시, 직선에서 멀어질수록 다른 색 표시
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['소계'],
           c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
             df_sort.index[n], fontsize=15)

plt.xlabel('인구수')
plt.ylabel('인구당비율')
plt.colorbar()
plt.grid()
plt.show()
```
